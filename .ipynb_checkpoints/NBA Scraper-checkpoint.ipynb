```python
import time
import bs4 as bs
import pandas as pd 

from urllib.request import FancyURLopener  
from random import choice 
from bs4 import SoupStrainer, BeautifulSoup

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('/usr/local/bin/chromedriver', chrome_options = chrome_options)

user_agents = [
    'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',
    'Opera/9.80 (X11; Linux i686; Ubuntu/14.10) Presto/2.12.388 Version/12.16',
    'Mozilla/5.0 (Windows; U; Windows NT 6.1; rv:2.2) Gecko/20110201',
    'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.75.14 (KHTML, like Gecko) Version/7.0.3 Safari/7046A194A',
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246'
]

class MyOpener(FancyURLopener, object):
    version = choice(user_agents)
myopener = MyOpener()

invidivual_url = 'https://www.basketball-reference.com/'

col_names = ['Position', 'Age', 'Team','Games', 'Games Started', 'Minutes Played',
             'Field Goals', 'Field Goals Attempted', 'Field Goal Percentage', 'Three Pointers', 
             'Three Pointers Attempted', 'Three Point Percentage', 'Two Pointers', 'Two Pointers Attemped',
             'Two Point Percentage', 'eFG%', 'Free Throws', 'Free Throws Attempted', 'Free Throw Percentage',
             'Offensive Rebounds', 'Defensive Rebounds', 'Total Rebound Percentage', 'Assists',
             'Steals', 'Blocks', 'Turnovers', 'Personal Fouls', 'Points']

# Clear lists by running this block
i_season_years = []
i_stat_names = []
i_stats_list = []

player_names_list = []
player_stats_list = []
player_stats_fixed = []

driver= webdriver.Chrome('/usr/local/bin/chromedriver', options = chrome_options)

######


which_stats = input('Would you like to search for invidivual player stats (i) or league wide stats (l): ')
if which_stats == 'i':
    individual_stat_scaper(individual_url)
    stat_lookup()
elif which_stats == 'l':
    which_season = input('Enter the season you would like to find input for (i.e. 2018-2019): ')
    league_scraper()
    

def individual_stat_scaper(url):
    driver.get(url)
    time.sleep(5)
    
    search_query = input('Enter the name of the player you would like to find stats for: ')
    search = driver.find_element_by_name('search')
    search.send_keys(search_query)
    submit = driver.find_element_by_css_selector("input[type='submit']")
    submit.send_keys(Keys.RETURN)
    time.sleep(5)
    player_url = str(driver.current_url)
    page = myopener.open(player_url)
    html = page.read().decode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')
    season = input('Enter the season you would like to see the stats for: ')
    if season:
        stats = soup.find_all('tr', {'id': 'per_game.' + season})
        if len(stats) == 0:
            print('You have probably selected a non-existent player, try again.')
    driver.quit()

        
def stat_lookup():
    stat_lines = stats[0].find_all('td', {'class': 'right'})
    player_age = stats[0].find('td', {'class': 'center'})

    stat_query = input('Enter the stat you would like to search for, press enter for all stats: ')
    if stat_query:
        for i in range(len(col_names)):
            stat = ''
            if stat_query in col_names[i]:
                stat = stat_lines[col_names.index(stat_query)].get_text()
                print(stat)
                i_stats_list.append(stat)
                i_stat_names.append(stat_query)
                i_season_years.append(str(int(season) - 1) + '-' + season)
                break
        if stat == '':
            print('You have entered an invalid stat, choose from the list below:')
            print(col_names)
    else:
        for line in stat_lines:
            print(line.get_text())
    continue_query = input('Would you like to find another stat? y or n: ')
    if continue_query == 'y':
        stat_lookup()
    else:
        pass 
    
def league_scraper():
    page = myopener.open('https://www.basketball-reference.com/leagues/NBA_' + which_season[5:9] + '_per_game.html')
    time.sleep(5)
    html = page.read().decode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')
    
    player_names_html = soup.find_all('td', {'class': 'left'})
    for i in range(0, len(player_names_html)):
        player_names = player_names_html[i].find_all('a')
        if 'player' in str(player_names):
            player_names_list.append(player_names[0].get_text())

    all_players = soup.find_all('tr')
    for i in range(1, len(all_players)):
        player = all_players[i].find_all('td')
        for j in range(1, len(player)):
            player_stats = player[j].get_text()
            player_stats_list.append(player_stats)
            
    x = 0
    y = 28
    for i in range(0,len(player_names_list)):
        player_stats_fixed.append(player_stats_list[x:y])
        x = x + 28
        y = y + 28
```

```python
len(player_stats_fixed)
```

```python
df = pd.DataFrame(player_stats_fixed, columns = col_names)
df.insert(0, 'Name', player_names_list)
df.set_index('Name')
df.astype({'Points': 'float'})
```

```python
filename = which_season+' Player Stats.csv'
df.to_csv(filename, sep='\t', index=False)
```
